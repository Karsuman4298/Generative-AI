```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification")
```

```python
classifier=pipeline("sentiment-analysis")
```

# Sentiment analysis

```python
result=classifier("i was not feel good yesterday")
```

```python
print(result)
```

```python
pipeline(task="sentiment-analysis")\
("‘Dhurandhar: The Revenge’ is being praised for its intense storytelling \
and strong detailing by Aditya Dhar.\
The film is described as impactful and engaging, with fans\
 calling it a must-watch and highlighting its powerful\
 presentation.")
```

# Mix Sentiment Analysis

```python
classifier=pipeline("sentiment-analysis",model="SamLowe/roberta-base-go_emotions")
```

```python

task_list=["After watching the film, I was lonely.\
",\
"My son was trying on his suit for prom, and he looked sharp, shaggy and shy all at once.",\
"Good news and no one to share it with."]

classifier(task_list)
```

# Text Generation


```python
text_generator=pipeline("text-generation",model="gpt2")
```

```python
generated_text=text_generator("what is AI",
                              truncation=True,
                              num_return_sequences=2)


print(generated_text[0]['generated_text'])
```
